# Compare HTML to MD extraction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv('../outputs/full_results.csv')
```

```python
results = results[(results.model_name == 'gpt-4o-mini-2024-07-18') & (results.n_studies == 156)]
```

```python
results
```

```python
sns.catplot(data=results, x='variable', y='avg_mean_percentage_error', kind='bar', hue='source', col='model_name', legend='full')
plt.xticks(rotation=20)
```

```python
sns.catplot(data=results, x='variable', y='summed_mean_percentage_error', kind='bar', hue='source', col='model_name', legend='full')
plt.xticks(rotation=20)
```

```python
sns.catplot(data=results, x='variable', y='hungarian_matched_error', kind='bar', hue='source', col='model_name', legend='full')
plt.xticks(rotation=20)
```

Using GPT 4o mini, HTML inputs yield very comparable accuracy to markdown inputs
